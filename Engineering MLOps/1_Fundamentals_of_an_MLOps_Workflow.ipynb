## L'évolution des infrastructures et du développement de logiciels

Avec la genèse de l'ère Internet moderne (vers 1995), nous avons assisté à une augmentation des applications logicielles, allant des systèmes d'exploitation tels que Windows 95 au système d'exploitation Linux et aux sites Web tels que Google et Amazon, qui servent le monde (en ligne) depuis plus de deux décennies. 

Cela a entraîné une culture d'amélioration continue des services en collectant, stockant et traitant une quantité massive de données provenant des interactions des utilisateurs. De tels développements ont façonné l'évolution de l'infrastructure informatique et le développement des logiciels. 

La transformation de l'infrastructure informatique s'est accélérée depuis le début de ce millénaire. Depuis lors, les entreprises ont de plus en plus adopté le cloud computing, car il ouvre de nouvelles possibilités pour les entreprises d'externaliser la maintenance de l'infrastructure informatique tout en fournissant les ressources informatiques nécessaires telles que les ressources de stockage et de calcul et les services nécessaires pour exécuter et faire évoluer leurs opérations. 

Le cloud computing offre un provisionnement à la demande et la disponibilité des ressources informatiques telles que le stockage de données et les ressources informatiques sans avoir besoin d'une gestion active par l'utilisateur des ressources informatiques. Par exemple, les entreprises qui approvisionnent les ressources de calcul et de stockage n'ont pas à gérer ces ressources directement et ne sont pas responsables de leur bon fonctionnement - la maintenance est sous-traitée au fournisseur de services cloud. 


Les entreprises utilisant le cloud computing peuvent en retirer des avantages car il n'est pas nécessaire d'acheter et de maintenir des ressources informatiques ; cela leur permet d'avoir moins d'expertise en interne pour la maintenance des ressources informatiques et cela permet aux entreprises d'optimiser les coûts et les ressources. Le cloud computing permet une mise à l'échelle à la demande et les utilisateurs paient en fonction de l'utilisation des ressources. En conséquence, nous avons vu des entreprises adopter le cloud computing dans le cadre de leurs activités et infrastructures informatiques.

Le cloud computing est devenu populaire dans l'industrie à partir de 2006, lorsque Sun Microsystems a lancé Sun Grid en mars 2006. Il s'agit d'un service de partage de ressources matérielles et de données. 

Ce service a été acquis par Oracle et a ensuite été nommé Sun Cloud. Parallèlement, la même année (2006), un autre service de cloud computing a été lancé par Amazon appelé Elastic Compute Cloud. Cela offre aux entreprises de nouvelles possibilités de provisionner des capacités de calcul, de stockage et de mise à l'échelle à la demande. 


Depuis lors, la transformation à travers les industries a été organique vers l'adoption du cloud computing. Au cours de la dernière décennie, de nombreuses entreprises à l'échelle mondiale et régionale ont catalysé la transformation du cloud, avec des entreprises telles que Google, IBM, Microsoft, UpCloud, Alibaba et d'autres investissant massivement dans la recherche et le développement de services cloud. 

En conséquence, un passage de l'informatique localisée (les entreprises ayant leurs propres serveurs et centres de données) à l'informatique à la demande a eu lieu en raison de la disponibilité de services cloud robustes et évolutifs. Désormais, les entreprises et les organisations sont en mesure de fournir des ressources à la demande sur le cloud pour répondre à leurs besoins de traitement de données. Avec ces développements, nous avons assisté à l'application de la loi de Moore, qui stipule que le nombre de transistors sur une puce double tous les 2 ans – bien que le coût des ordinateurs ait diminué de moitié, cela a été vrai jusqu'à présent. Par la suite, certaines tendances se développent comme suit

## La fin de la loi de Moore

Avant 2012, les résultats de l'IA suivaient de près la loi de Moore, le calcul doublant tous les 2 ans. Après 2012, le calcul a doublé tous les 3,4 mois (provenant de AI Index 2019 – https://hai.stanford.edu/research/ai-index-2019). Nous pouvons observer à partir de la figure 1.1 que la demande d'apprentissage en profondeur et de calcul haute performance (HPC) a augmenté de façon exponentielle avec une croissance d'environ 35 fois de l'informatique tous les 18 mois alors que la loi de Moore est considérée comme dépassée (2 fois tous les 18 mois). La loi de Moore est toujours applicable au cas des CPU (performances monocœur) mais pas aux nouvelles architectures matérielles telles que les GPU et les TPU. Cela rend la loi de Moore obsolète et dépassée contrairement aux demandes et tendances actuelles.

## Évolution du développement logiciel 

Le développement logiciel a évolué de pair avec les développements d'infrastructure pour faciliter le développement d'applications efficace utilisant l'infrastructure. Le modèle en cascade présente de nombreuses limitations, qui ont conduit à l'évolution du développement de logiciels au fil des ans sous la forme de méthodologies Agile et de la méthode DevOps, comme le montre la figure 1.3 :

## La méthode de la cascade 

La méthode de la cascade (waterfall) a été utilisée pour développer des logiciels dès le début de l'ère d'Internet (~1995). C'est une manière non itérative de développer un logiciel. Il est délivré de manière unidirectionnelle. Chaque étape est pré-organisée et exécutée les unes après les autres, depuis la collecte des exigences jusqu'à la conception, le développement et les tests du logiciel. 

La méthode en cascade est faisable et appropriée lorsque les exigences sont bien définies, spécifiques et ne changent pas avec le temps. Par conséquent, cela ne convient pas aux projets dynamiques où les exigences changent et évoluent selon les demandes des utilisateurs. Dans de tels cas, où il y a une modification continue, la méthode en cascade ne peut pas être utilisée pour développer un logiciel. Voici les principaux inconvénients des méthodes de développement en cascade : 

* L'ensemble des exigences doit être fourni avant de commencer le développement ; les modifier pendant ou après le développement du projet n'est pas possible.

* Il y a moins de chances de créer ou de mettre en œuvre des composants réutilisables.

* Les tests ne peuvent être effectués qu'une fois le développement terminé. Les tests ne sont pas destinés à être itérables ; il n'est pas possible de revenir en arrière et de réparer quoi que ce soit une fois que c'est fait. De plus, les tests d'acceptation des clients introduisaient souvent des changements, entraînant un retard de livraison et des coûts élevés. Ce mode de développement et de test peut avoir un impact négatif sur le calendrier et les coûts de livraison du projet.

* La plupart du temps, les utilisateurs du système disposent d'un système basé sur la compréhension du développeur, qui n'est pas centré sur l'utilisateur et peut venir à court de répondre à leurs besoins

## La méthode Agile

La méthode Agile facilite une approche itérative et progressive du développement de logiciels. Contrairement à la méthode en cascade, les approches agiles sont précises et centrées sur l'utilisateur. 

La méthode est bidirectionnelle et implique souvent les utilisateurs finaux ou les clients dans le processus de développement et de test afin qu'ils aient la possibilité de tester, de donner leur avis et de suggérer des améliorations tout au long du processus et des phases de développement du projet. 

Agile présente plusieurs avantages par rapport à la méthode en cascade :

* Les exigences sont définies avant de commencer le développement, mais elles peuvent être modiées à tout moment.

* Il est possible de créer ou de mettre en œuvre des composants réutilisables.

* La solution ou le projet peut être modulaire en séparant le projet en différents modules qui sont livrés périodiquement.

* Les utilisateurs ou les clients peuvent co-créer en testant et en évaluant périodiquement les modules de solution développés pour s'assurer que les besoins de l'entreprise sont satisfaits. Un tel processus centré sur l'utilisateur garantit des résultats de qualité axés sur la satisfaction des besoins des clients et des entreprises.

## La méthode DevOps 

La méthode DevOps étend les pratiques de développement agile en rationalisant davantage le mouvement du changement de logiciel à travers les étapes de construction, de test, de déploiement et de livraison. DevOps donne aux équipes interfonctionnelles l'autonomie nécessaire pour exécuter leurs applications logicielles basées sur l'intégration continue, le déploiement continu et la livraison continue. Il encourage la collaboration, l'intégration et l'automatisation entre les développeurs de logiciels et les opérateurs informatiques afin d'améliorer l'efficacité, la vitesse et la qualité de la fourniture de logiciels centrés sur le client. 

DevOps fournit un cadre de développement logiciel rationalisé pour la conception, le test, le déploiement et la surveillance des systèmes en production. DevOps a permis d'expédier le logiciel à la production en quelques minutes et de le faire fonctionner de manière fiable

## Défis du développement de logiciels traditionnels

Dans la section précédente, nous avons observé le changement dans le développement de logiciels traditionnels du modèle en cascade aux pratiques agiles et DevOps. 

Les pratiques Agile et DevOps ont permis aux entreprises de livrer des logiciels de manière fiable. DevOps a permis de livrer le logiciel à la production en quelques minutes et de le maintenir en fonctionnement fiable. Cette approche a connu un tel succès que de nombreuses entreprises l'adoptent déjà, alors pourquoi ne pouvons-nous pas continuer à faire la même chose pour les applications ML ? 

La principale cause est qu'il existe une différence fondamentale entre le développement ML et les logiciels traditionnels développement :

L'apprentissage automatique n'est pas seulement du code ; c'est du code plus des données. Un modèle ML est créé en appliquant un algorithme (via un code) pour adapter les données afin d'aboutir à un modèle ML, comme le montre la figure 1.5

image.png

Alors que le code est méticuleusement créé dans l'environnement de développement, les données proviennent de plusieurs sources pour la formation, les tests et l'inférence. Il est robuste et change dans le temps en termes de volume, de vélocité, de véracité et de variété. Pour suivre l'évolution des données, le code évolue au fil du temps. Pour la perspective, leur relation peut être observée comme si le code et les données vivaient dans des plans séparés qui partagent la dimension temporelle mais sont indépendants dans tous les autres aspects. L'enjeu d'un processus de développement ML est de créer un pont entre ces deux plans de manière contrôlée

image.png

Les données et le code, avec la progression du temps, finissent par aller dans deux directions avec un objectif de construire et de maintenir un système de ML robuste et évolutif. Cette déconnexion pose plusieurs défis qui doivent être résolus par quiconque essaie de mettre un modèle ML en production. Cela s'accompagne de défis tels qu'un déploiement lent, fragile, fragmenté et incohérent, ainsi qu'un manque de reproductibilité et de traçabilité.


Pour surmonter ces défis, MLOps propose une approche systématique en reliant les données et le code au fil du temps. C'est la solution aux défis posés par les méthodes traditionnelles de développement de logiciels en ce qui concerne les applications ML. En utilisant la méthode MLOps, les données et le code progressent au fil du temps dans une direction avec un objectif de créer et de maintenir un système de ML robuste et évolutif :

image.png

MLOps facilite le développement, le déploiement et la surveillance des modèles de ML dans une approche rationalisée et systématique. Il permet aux équipes de science des données et informatiques de collaborer, de valider et de régir leurs opérations. Toutes les opérations exécutées par les équipes sont enregistrées ou auditées, traçables de bout en bout et reproductibles. Dans les sections à venir, nous apprendrons comment MLOps permet aux équipes de science des données et informatiques de créer et de maintenir des systèmes de ML robustes et évolutifs.

## Concepts et flux de travail des MLOps 

Dans cette section, nous allons découvrir un flux de travail MLOps générique ; c'est le résultat de nombreuses itérations du cycle de conception, comme discuté dans la section précédente. Ce flux de travail est segmenté en deux modules :

* Pipeline MLOps (construire, déployer et surveiller) - la couche supérieure

* Drivers : données, code, artefacts, middleware et infrastructure - couches intermédiaire et inférieure


La couche supérieure est le pipeline MLOps (construire, déployer et surveiller), qui est activé par des pilotes tels que les données, le code, les artefacts, le middleware et l'infrastructure. Le pipeline MLOps est alimenté par une gamme de services, de pilotes, de middleware et d'infrastructure, et il crée des solutions basées sur le ML. 

En utilisant ce pipeline, une entreprise ou un ou plusieurs individus peuvent effectuer un prototypage, des tests et une validation rapides et déployer le ou les modèles en production à grande échelle de manière frugale et efficace. ## Le pipeline MLOps

Le pipeline MLOps est la couche supérieure, qui effectue des opérations telles que la construction, le déploiement et la surveillance, qui fonctionnent de manière modulaire et synchronisées les unes avec les autres. Examinons les fonctionnalités de chaque module.

## build

Le module de build est le pipeline ML de base, et c'est uniquement pour la training, et le packaging et la gestion des versions des modèles ML. Il est alimenté par les ressources de calcul requises (par exemple, le CPU ou le GPU sur le cloud ou l'informatique distribuée) pour exécuter la formation et le pipeline de ML :

image.png

Le pipeline fonctionne de gauche à droite. Examinons en détail la fonctionnalité de chaque étape : 

* Ingestion de données : cette étape est une étape de déclenchement pour le pipeline de ML. Il traite le volume, la vitesse, la véracité et la variété des données en extrayant des données de diverses sources de données (par exemple, des bases de données, des entrepôts de données ou des lacs de données) et en ingérant les données requises pour l'étape d'apprentissage du modèle. 
Des pipelines de données robustes connectés à plusieurs sources de données lui permettent d'effectuer des opérations d'extraction, de transformation et de chargement (ETL) pour fournir les données nécessaires à des fins de formation ML. Dans cette étape, nous pouvons diviser et versionner les données pour l'apprentissage du modèle dans le format requis (par exemple, l'ensemble d'apprentissage ou de test). Comme les modèles ML génèrent des données en production pendant l'inférence, ces données peuvent être liées à la formation du modèle et à la lignée de déploiement pour garantir la lignée de bout en bout, ce qui est important pour certaines exigences de conformité. ## Drivers

Ce sont les principaux moteurs du pipeline MLOps : données, code, artefacts, middleware et infrastructure. Examinons chacun des pilotes pour avoir un aperçu de la façon dont ils activent le pipeline MLOps :


image.png

Chacun des principaux moteurs du pipeline MLOps est défini comme suit :

* data : Les données peuvent se présenter sous plusieurs formes, telles que du texte, de l'audio, de la vidéo et des images. Dans les applications logicielles traditionnelles, les données ont souvent tendance à être structurées, alors que, pour les applications ML, elles peuvent être structurées ou non structurées. Pour gérer les données dans les applications ML, les données sont gérées au cours de ces étapes : acquisition de données, annotation de données, catalogage des données, préparation des données, vérification de la qualité des données, échantillonnage des données et augmentation des données. Chaque étape implique son propre cycle de vie. Cela crée un tout nouvel ensemble de processus et d'outils nécessaires aux applications ML. Pour un fonctionnement efficace du pipeline ML, les données sont segmentées et versionnées en données d'entraînement, données de test et données de surveillance (collectées en production, par exemple, entrées de modèle, sorties et données de télémétrie). Ces opérations de données font partie du pipeline MLOps

* Code : il existe trois modules de code essentiels qui pilotent le pipeline MLOps : le code d'entraînement, le code de test et le code d'application. Ces scripts ou codes sont exécutés à l'aide du CI/CD et des pipelines de données pour garantir le bon fonctionnement du pipeline MLOps. Le système de gestion du code source (par exemple, en utilisant Git ou Mercurial) permettra l'orchestration et jouera un rôle essentiel dans la gestion et l'intégration transparente avec CI, CD et les pipelines de données. Tout le code est mis en scène et versionné dans la configuration de gestion du code source (par exemple, Git)

* Artefacts : le pipeline MLOps génère des artefacts tels que des données, des modèles sérialisés, des extraits de code, des journaux système, la formation de modèle ML et des informations sur les métriques de test. Tous ces artefacts sont utiles au bon fonctionnement du pipeline MLOps, garantissant sa traçabilité et sa durabilité. Ces artefacts sont gérés à l'aide de services middleware tels que le registre de modèles, les espaces de travail, les services de journalisation, les services de gestion du code source, les bases de données, etc. .


* Middleware : Le middleware est un logiciel informatique qui offre des services aux applications logicielles qui sont plus que ceux disponibles à partir des systèmes d'exploitation. Les services middleware garantissent plusieurs applications pour automatiser et orchestrer les processus pour le pipeline MLOps. Nous pouvons utiliser un ensemble diversifié de logiciels et de services middleware en fonction des cas d'utilisation, par exemple, Git pour la gestion du code source, des réseaux virtuels pour activer les configurations réseau requises, Docker pour la conteneurisation de nos modèles et Kubernetes pour l'orchestration des conteneurs pour automatisez le déploiement, la mise à l'échelle et la gestion des applications.

* Infrastructure : Pour garantir le bon fonctionnement du pipeline MLOps, nous avons besoin de ressources de calcul et de stockage essentielles pour former, tester et déployer les modèles de ML. Les ressources de calcul nous permettent de former, déployer et surveiller nos modèles de ML. Deux types de ressources de stockage peuvent faciliter les opérations de ML, le stockage central et les magasins de fonctionnalités. Un stockage central stocke les journaux, les artefacts, les données de formation, de test et de surveillance. Un magasin de fonctionnalités est facultatif et complémentaire au stockage central. Il extrait, transforme et stocke les fonctionnalités nécessaires à la formation et à l'inférence du modèle ML à l'aide d'un pipeline de fonctionnalités. En ce qui concerne l'infrastructure, il existe diverses options telles que les ressources sur site ou l'infrastructure en tant que service (IaaS), qui sont des services cloud. De nos jours, de nombreux acteurs du cloud proposent l'IaaS, tels qu'Amazon, Microso, Google, Alibaba, etc. Avoir la bonne infrastructure pour votre cas d'utilisation permettra des opérations robustes, efficaces et économes pour votre équipe et votre entreprise. 

Un flux de travail entièrement automatisé est réalisable avec une optimisation intelligente et une synergie de tous ces pilotes avec le pipeline MLOps. Certains avantages directs de la mise en œuvre d'un flux de travail MLOps automatisé sont un pic d'efficacité des équipes informatiques (en réduisant le temps passé par les scientifiques des données et les développeurs sur des tâches banales et répétables) et l'optimisation des ressources, entraînant des réductions de coûts, et ces deux sont parfaits pour toute entreprise.
