## Validation croisée 

L'évaluation d'un modèle de ML est vitale pour comprendre son comportement et cela peut s'avérer délicat. Normalement, l'ensemble de données est divisé en deux sous-ensembles : les ensembles d'apprentissage et de test. Tout d'abord, l'ensemble d'apprentissage est utilisé pour entraîner le modèle, puis l'ensemble de test est utilisé pour tester le modèle. Après cela, les performances du modèle sont évaluées pour déterminer l'erreur à l'aide de métriques telles que le pourcentage de précision du modèle sur les données de test. Cette méthodologie n'est ni fiable ni complète car la précision d'un ensemble de tests peut différer de celle d'un autre ensemble de tests. Pour éviter ce problème, la validation croisée fournit une solution en fragmentant ou en divisant l'ensemble de données en plis et en s'assurant que chaque pli est utilisé comme jeu de test à un moment donné, comme le montre la figure 5.2 :


image.png

Il existe plusieurs méthodes de validation croisée, y compris la validation croisée stratifiée, la validation croisée Leave-one-out et la validation croisée K-fold. La validation croisée à K plis est largement utilisée et mérite d'être notée car cette technique consiste à diviser l'ensemble de données en k plis/fragments, puis à utiliser chaque pli comme jeu de test au cours d'itérations successives. 


Ce processus est utile car chaque itération possède un ensemble de tests unique sur lequel la précision est mesurée. Ensuite, la précision de chaque itération est utilisée pour trouver les résultats moyens des tests (calculés en prenant simplement la moyenne de tous les résultats des tests). 

La précision moyenne acquise par validation croisée est une métrique plus fiable et plus complète que la mesure de précision conventionnelle. Par exemple, sur la figure 5.2, nous pouvons voir cinq itérations. Chacune de ces itérations a un ensemble de tests unique, et lors du test de précision pour chaque itération et en faisant la moyenne de toutes les précisions, nous obtenons une précision moyenne pour le modèle en utilisant la validation croisée K-fold. Il convient de noter que K-fold n'est pas un bon choix si vous disposez d'un très grand ensemble de données d'entraînement ou si le modèle nécessite une grande quantité de temps, de traitement CPU et/ou GPU pour l'exécution.

## Précision

Lorsqu'un classificateur est formé, la précision peut être une mesure vitale pour quantifier les prédictions de classe positives faites par le classificateur qui sont réellement vraies et appartiennent à la classe positive. La précision quantifie le nombre de prédictions positives correctes. Par exemple, disons que nous avons formé un classificateur pour prédire les chats et les chiens à partir d'images. En inférant le modèle entraîné sur les images de test, le modèle est utilisé pour prédire/détecter les chiens à partir des images (en d'autres termes, les chiens étant la classe positive). La précision, dans ce cas, quantifie le nombre de prédictions de chien correctes (prédictions positives). 

La précision est calculée comme le rapport des exemples positifs correctement prédits au nombre total d'exemples positifs prédits. Précision = TruePositives / (TruePositives + FalsePositives)

Précision se concentre sur la réduction des faux positifs. La haute précision va de 0 à 1, et se rapporte à un faible taux de faux positifs. Plus la précision est élevée, mieux c'est ; par exemple, un modèle de classeur d'images qui prédit si un patient cancéreux a besoin d'un traitement de chimiothérapie. Si le modèle prédit qu'un patient doit être soumis à une chimiothérapie alors que cela n'est pas vraiment nécessaire, cela peut être très néfaste car les eeffts de la chimiothérapie peuvent être néfastes lorsqu'elle n'est pas requise. Ce cas est un faux positif dangereux. Un score de haute précision entraînera moins de faux positifs, tandis qu'un score de faible précision entraînera un nombre élevé de faux positifs. Par conséquent, en ce qui concerne le traitement de chimiothérapie, le modèle de prédiction doit avoir un score de haute précision

## Rappel

Lorsqu'un classificateur est entraîné, le rappel peut être utilisé pour quantifier les prédictions de classe positives établies à partir du nombre total d'exemples positifs dans l'ensemble de données. Le rappel mesure le nombre de prédictions positives correctes faites sur le nombre total de prédictions positives qui auraient pu être faites. Le rappel fournit des preuves de prédictions positives manquées, contrairement à la précision, qui ne nous indique que les prédictions positives correctes sur le nombre total de prédictions positives. à partir d'images. En inférant le modèle entraîné sur les images de test, le modèle est utilisé pour prédire/détecter les chiens à partir des images (en d'autres termes, les chiens étant la classe positive). Le rappel, dans ce cas, quantifie le nombre de prédictions de chien manquées (prédictions positives). De cette manière, le rappel fournit une indication empirique de la couverture de la classe positive. Rappel = TruePositives / (TruePositives + FalseNegatives)

Le rappel se concentre sur la minimisation faux négatifs. Un rappel élevé correspond à un faible taux de faux négatifs. Plus le rappel est élevé, mieux c'est. Par exemple, un modèle qui analyse les données de prol d'un passager dans un aéroport essaie de prédire si ce passager est un terroriste potentiel. Dans ce cas, il est plus sûr d'avoir des faux positifs que des faux négatifs. Si les modèles prédisent qu'une personne innocente est un terroriste, cela pourrait être vérifié à la suite d'une enquête plus approfondie. Mais si un terroriste passe, plusieurs vies pourraient être en danger. Dans ce cas, il est plus sûr d'avoir des faux négatifs que des faux positifs car les faux négatifs peuvent être vérifiés à l'aide d'une enquête approfondie. Le rappel doit être élevé pour éviter les faux négatifs. Dans ce cas, avoir un score de rappel élevé est prioritaire sur une haute précision


## F-score 

Dans un cas où nous devons éviter à la fois des faux positifs élevés et des faux négatifs élevés, le score f est une mesure utile pour atteindre cet état. La mesure F fournit un moyen de consolider à la fois la précision et le rappel en une seule métrique qui reflète les deux propriétés. Ni la précision ni le rappel ne décrivent toute l'histoire. Nous pouvons avoir la meilleure précision avec un rappel terrible, ou alternativement, la mesure F exprime à la fois la précision et rappel. Il est mesuré selon la formule suivante :

Mesure F = (2 * Précision * Rappel) / (Précision + Rappel)

la moyenne harmonique de votre précision et de votre rappel est la F-mesure. Dans la plupart des cas, vous devez choisir entre précision et rappel. La moyenne harmonique diminue rapidement si vous optimisez votre classificateur pour favoriser l'un et défavoriser l'autre. Lorsque la précision et le rappel sont similaires, c'est à son meilleur ; par exemple, un modèle qui prédit le cancer à un stade précoce en prenant en entrée les images et les examens sanguins d'un patient. 

Dans ce scénario réel, cela pourrait entraîner de nombreux coûts inutiles pour un hôpital et nuire à la santé d'un patient si le modèle génère un nombre élevé de faux positifs. D'un autre côté, si le modèle ne parvient pas à détecter les vrais patients atteints de cancer, plusieurs vies seraient en danger. Dans de tels cas, nous devons éviter à la fois les faux positifs et les faux négatifs élevés et ici, le score f est une mesure utile pour éviter les faux positifs et les faux négatifs élevés. Le score F mesure entre 0 et 1. Plus le score f est élevé, mieux c'est. On peut s'attendre à un plus petit nombre de faux positifs et de faux négatifs avec un f-score élevé

## Matrice de confusion 

La matrice de confusion est une métrique qui rapporte les performances des modèles de classification sur un ensemble d'échantillons de données de test pour lesquels les valeurs de prédiction sont pré-connues. Il s'agit d'une métrique sous forme de matrice où une matrice de confusion est une matrice N X N et N est le nombre de classes prédites. Par exemple, disons que nous avons deux classes à prédire (classication binaire), alors N=2, et, par conséquent, nous aurons une matrice 2 X 2, comme celle montrée ici sur la figure 5.3 :

Il y a 181 échantillons de données de test sur lesquels des prédictions sont faites pour classer les échantillons de données de patients en catégories diabétiques et non diabétiques. En utilisant une matrice de confusion, vous pouvez obtenir des informations critiques pour interpréter les performances du modèle. Par exemple, en un coup d'œil, vous saurez combien de prédictions faites sont réellement vraies et combien sont de faux positifs. De telles informations sont inestimables pour interpréter les performances du modèle dans de nombreux cas. Voici ce que signifient ces termes dans le contexte de la matrice de confusion : 

* Vrais positifs (TP) : ce sont des cas prédits comme étant oui et sont en fait oui selon les échantillons de données de test. 

* Vrais négatifs (VN) : ce sont les cas prédits comme non et ces cas sont en fait non selon les échantillons de données de test. 

* Faux positifs (FP) : le modèle a prédit oui, mais ils sont non selon les échantillons de données de test. Ce type d'erreur est connu sous le nom d'erreur de type I. 

* Faux négatifs (FN) : le modèle a prédit non, mais ils sont oui selon les échantillons de données de test. Ce type d'erreur est appelé erreur de type II.

La matrice de confusion est la métrique d'analyse d'erreur de facto pour les problèmes de classification, car la plupart des autres métriques sont dérivées de celle-ci

## AUC-ROC

Une perspective différente pour observer les performances du modèle peut nous permettre d'interpréter les performances du modèle et de l'affiner pour obtenir de meilleurs résultats. Les courbes ROC et AUC peuvent permettre de telles informations. Voyons comment la courbe des caractéristiques de fonctionnement du récepteur (ROC) peut nous permettre d'interpréter les performances du modèle. 

La courbe ROC est un graphique présentant les performances d'un modèle de classification à tous les seuils de classification. Le graphique utilise deux paramètres pour décrire les performances du modèle : le taux de vrais positifs (TPR=TP/TP+FN) et le taux de faux positifs (FPR=FP/FP+TN). Le diagramme suivant montre une courbe ROC typique.

Une courbe ROC représente le TPR par rapport au FPR pour différents seuils de classification. L'abaissement du seuil de classification permet à davantage d'éléments d'être classés comme positifs, ce qui à son tour augmente à la fois les faux positifs et les vrais positifs. L'aire sous la courbe (AUC) est une métrique utilisée pour quantifier l'efficacité ou la capacité d'un classificateur à distinguer les classes et est utilisée pour résumer la courbe ROC


La valeur AUC varie de 0 à 1, et le classificateur est capable de distinguer correctement tous les points de classe positifs et négatifs si la valeur AUC est de 1, et le classificateur est incapable de distinguer correctement tous les points de classe positifs et négatifs si la valeur AUC est 0. Lorsque la valeur AUC est de 0,5 (sans définir manuellement un seuil), alors il s'agit d'un classificateur aléatoire.


L'AUC nous aide à classer les prédictions en fonction de leur précision, mais elle ne nous donne pas de valeurs absolues. Il est donc indépendant de l'échelle. De plus, l'AUC est indépendante du seuil de classification. Le seuil de classification choisi n'a pas d'importance lors de l'utilisation de l'AUC car l'AUC estime la qualité des prédictions du modèle quel que soit le seuil de classification choisi

## Le coefficient de corrélation de Matthew

Brian Matthews a développé le coefficient de corrélation de Matthews (MCC) en 1975 comme méthode d'évaluation de modèle. Il calcule les écarts entre les valeurs réelles et attendues. C'est une extension des résultats de la matrice de confusion pour mesurer l'inefficacité d'un classificateur. TP, TN, FP et FN sont les quatre entrées d'une matrice de confusion. Ces entrées sont prises en compte dans le coefficient :

image.png

Cette mesure ne donne des scores élevés que lorsqu'une prédiction renvoie de bons taux pour toutes ces quatre catégories. Le score MCC varie de -1 à +1 :

* 1 est le meilleur accord entre les données réelles et les prévisions.
* Lorsque le score est de 0, cela signifie qu'il n'y a aucune concordance entre les données réelles et les prévisions. 

La prédiction est aléatoire par rapport aux données réelles. Par exemple, un score MCC de 0,12 suggère que le classificateur est très aléatoire. S'il est de 0,93, cela suggère que le classificateur est bon. Le MCC est une métrique utile pour aider à mesurer l'inefficacité d'un classificateur.

## Le Rand index

Le Rand index est une métrique permettant d'évaluer la qualité de la technique de clustering. Il décrit le degré de similitude entre les clusters. L'indice Rand mesure le pourcentage de bonnes décisions. Les décisions attribuent une paire de points de données (par exemple, des documents) au même cluster. Si N points de données existent, le nombre total de décisions = N(N-1)/2, qui désigne la paire de points de données impliqués dans la décision. Indice de Rand = TP + TN / TP + FP + FN + TN

## Purity

Purity est une métrique d'évaluation externe pour la qualité des clusters. Pour calculer la pureté, les clusters sont étiquetés en fonction de la classe la plus courante dans le cluster, puis la précision de cette affectation de cluster est mesurée en calculant le nombre de points de données correctement attribués et en divisant par N (nombre total de points de données regroupés). Un bon clustering a une valeur de pureté proche de 1, et un mauvais clustering a une valeur de pureté proche de 0. ## Le coefficient Silhouette

Pour les algorithmes de clustering, il est important de déterminer la qualité des clusters. Pour déterminer la qualité ou la qualité des clusters, le score de silhouette, ou coefficient de silhouette, est utilisé comme mesure. Sa valeur va de -1 à 1. Lorsque les clusters sont clairement distinguables ou bien séparés les uns des autres, alors le score de silhouette est de 1. Au contraire, -1 signifie que les clusters sont mal alloués, et 0 signifie que les clusters sont indifférents les uns aux autres. une autre. 

C'est ainsi que le score de silhouette est calculé : 

Score de silhouette = (ba)/max(a,b)

a = la distance moyenne entre chaque point au sein d'un cluster (distance intra-cluster moyenne).

b = la distance moyenne entre tous les clusters (distance moyenne inter-clusters).

## Test humain contre machine
Le zèle pour atteindre des performances de niveau humain est assez courant lors de la formation et du test des modèles de ML et d'apprentissage en profondeur. Afin de valider les modèles et de conclure que les modèles ont atteint ou dépassé les performances de niveau humain, des expériences humaines contre machines sont effectuées sur des tâches. La même tâche est mise en œuvre à l'aide d'un modèle ML et la performance humaine est évaluée par rapport à la performance du modèle ML. Il existe différentes métriques pour évaluer les performances de l'homme par rapport à la machine en fonction du contexte et des tâches. Quelques exemples sont mentionnés ici :

* L'évaluation bilingue sous-étudiée (BLEU) est une méthode d'évaluation de la qualité d'un texte pour la tâche de traduction automatique d'une langue à une autre. La qualité du texte généré par un algorithme de traduction automatique est comparée à la sortie d'un humain. L'évaluation est faite pour observer à quel point une traduction automatique est proche d'une traduction humaine professionnelle.


* Recall-Oriented Understudy for Gisting Evaluation (ROUGE) (ROUGE) est une mesure d'évaluation des performances de l'homme par rapport à la machine utilisée pour évaluer des tâches telles que la synthèse automatique et la traduction automatique. Cette métrique compare un résumé ou une traduction généré automatiquement avec un résumé/des traductions produites par des humains

Dans un autre sens, le test de Turing est également un test pour évaluer la capacité d'une machine à tromper un humain en lui faisant croire qu'une tâche effectuée par une machine est humaine ou effectuée par un humain. ## Récompense par retour (Reward per return)

Les modèles d'apprentissage par renforcement sont des modèles hybrides qui impliquent des mécanismes d'apprentissage continu entre l'agent et l'environnement d'exploitation afin d'atteindre des objectifs prédéfinis. L'agent apprend en fonction des récompenses obtenues pour des étapes ecaces ou optimales vers l'atteinte d'un objectif. Lorsque l'objectif est un contrôle optimal, vous souhaiterez mesurer l'agent en fonction de ses performances dans la tâche. Pour quantifier dans quelle mesure l'agent exécute la tâche, les mesures agrégées de la récompense, telles que la récompense totale par épisode (autrement appelée « retour ») ou la récompense moyenne par pas de temps, peuvent être utilisées pour évaluer et optimiser le contrôle de l'agent en ce qui concerne à l'environnement et aux objectifs

Le regret minimum est de 0 avec la politique optimale. Plus le regret est petit, meilleure est la performance d'un algorithme. ## MIMIC Explainer

Mimic Explainer est une approche imitant les modèles de boîte noire en formant un modèle de substitution global interprétable. Ces modèles de substitution globaux entraînés sont des modèles interprétables qui sont entraînés pour approximer les prédictions de tout modèle de boîte noire aussi précisément que possible. En utilisant le modèle de substitution, un modèle boîte noire peut être évalué ou interprété comme suit. 

Les étapes suivantes sont mises en œuvre pour former un modèle de substitution :


1. Pour entraîner un modèle de substitution, commencez par sélectionner un ensemble de données, X. Cet ensemble de données peut être le même que celui utilisé pour l'entraînement du modèle boîte noire ou il peut s'agir d'un autre ensemble de données de distributions similaires selon le cas d'utilisation.

2. Obtenez les prédictions du modèle boîte noire pour l'ensemble de données sélectionné, X.

3. Sélectionnez un type de modèle interprétable (modèle linéaire, arbre de décision, forêt aléatoire, etc.).

4. À l'aide de l'ensemble de données X et de ses prédictions, entraînez le modèle interprétable.

5. Vous avez maintenant un modèle de substitution formé. Bravo !

6. Évaluez dans quelle mesure le modèle de substitution a reproduit les prédictions du modèle boîte noire, par exemple, en utilisant R-carré ou F-score.

7. Obtenez une compréhension des prédictions du modèle boîte noire en interprétant le modèle de substitution. Les modèles interprétables suivants peuvent être utilisés comme modèles de substitution : modèle d'amplification du gradient de lumière (LightGBM), régression linéaire, descente de gradient stochastique ou forêt aléatoire et arbre de décision.

Surrogate Les modèles peuvent permettre aux développeurs de solutions de ML d'évaluer et de comprendre les performances du modèle de boîte noire.

## Permutation feature importance explainer (PFI)

L'importance des caractéristiques de permutation (PFI) est une technique utilisée pour expliquer les modèles de classification et de régression. Cette technique est utile pour interpréter et comprendre une caractéristique pour modéliser la sortie ou la corrélation de prédiction. Le PFI est une alternative au SHAP. Il fonctionne en évaluant au hasard une caractéristique à la fois pour l'ensemble de données et en calculant le changement dans les métriques d'évaluation des performances. Le changement dans la métrique de performance est évalué pour chaque caractéristique ; plus le changement est important, plus la caractéristique est importante. Le PFI peut décrire le comportement global de n'importe quel modèle, mais n'explique pas les prédictions individuelles du modèle. PFI est une alternative à SHAP, mais reste assez différent car PFI est basé sur la diminution des performances du modèle, tandis que SHAP est basé sur l'ampleur des attributions de fonctionnalités.

## La moyenne

la moyenne, est la valeur centrale de l'ensemble de données. Il est calculé en additionnant toutes les valeurs et en divisant la somme par le nombre de valeurs


## Écart-type : 

l'écart-type mesure la dispersion des valeurs dans l'ensemble de données. Plus l'écart type est faible, plus les données se rapprochent de la moyenne. Un ensemble de données largement répandu aurait un écart type plus élevé

## Bias

Bias mesure la force (ou la rigidité) de la fonction de mappage entre les variables indépendantes (entrée) et dépendantes (sortie). plus les hypothèses du modèle sur la forme fonctionnelle de la cartographie sont fortes, plus le biais est important

Un biais élevé est utile lorsque le modèle vrai (mais inconnu) sous-jacent a des propriétés correspondantes en tant qu'hypothèses de la fonction de mappage. Cependant, vous pourriez vous retrouver complètement hors de propos si le modèle sous-jacent ne présente pas des propriétés similaires à celles de la forme fonctionnelle du mappage. Par exemple, l'hypothèse qu'il existe une relation linéaire dans les variables alors qu'en réalité elle est fortement non linéaire et cela conduirait à un mauvais fit :

* Faible biais : hypothèses faibles en ce qui concerne la forme fonctionnelle de la cartographie des entrées aux extrants

* Biais élevé : hypothèses fortes concernant la forme fonctionnelle du mappage des intrants aux extrants


Le biais est toujours une valeur positive. Voici une ressource supplémentaire pour en savoir plus sur les biais dans le ML. Cet article propose une explication plus large : https://kourentzes.com/forecasting/2014/12/17/the-bias-coefficient-a-new-metric-for-forecast-bias/.


## Variancece 

La variance de le modèle est le degré auquel les performances du modèle changent lorsqu'il est ajusté sur différentes données d'entraînement. L'impact des spécificités sur le modèle est capturé par la variance. Un modèle à forte variance changera beaucoup, même avec de petits changements dans l'ensemble de données d'apprentissage. D'un autre côté, un modèle à faible variance ne changerait pas beaucoup, même avec des changements importants dans l'ensemble de données d'apprentissage. 

La variance est toujours positive. 


## R-carré

R-carré, également appelé coeficient de détermination, mesure la variation de la variable dépendante qui peut être expliquée par le modèle. Il est calculé comme la variation expliquée divisée par la variation totale. En termes simples, R-carré mesure à quel point les points de données sont proches de la droite de régression ajustée. La valeur de R-carré se situe entre 0 et 1. Une faible valeur de R-carré indique que la majeure partie de la variation de la variable de réponse n'est pas expliquée par le modèle, mais par d'autres facteurs qui n'y sont pas inclus. 

En général, vous devriez viser une valeur R carré plus élevée car cela indique que le modèle correspond mieux aux données. 

## RMSE : 

l'erreur quadratique moyenne (RMSE) mesure la différence entre les valeurs prédites du modèle et les valeurs observées. (vraies) valeurs.

Les options sont nombreuses et vous devez choisir la bonne métrique pour les scénarios de production du monde réel afin d'avoir des évaluations bien justifiées ; par exemple, pourquoi un data scientist ou une équipe de data science peut vouloir sélectionner une métrique d'évaluation plutôt qu'une autre, par exemple, R au carré sur la moyenne pour un problème de régression. Cela dépend du cas d'utilisation et du type de données

## Métriques du modèle HITL : 

il existe deux types de modèles HITL : l'apprentissage par renforcement humain et les modèles d'apprentissage actif. Dans ces modèles, la collaboration homme-machine encourage l'algorithme à imiter les comportements et les résultats humains. L'humain dans la boucle est l'un des principaux moteurs de ces solutions de ML. Les humains valident, étiquettent et recyclent les modèles pour maintenir la précision du modèle


## Biais humains

Tout comme le cerveau humain, les systèmes de ML sont sujets à des biais cognitifs. Les biais cognitifs humains sont des processus qui perturbent votre prise de décision et votre capacité de raisonnement, aboutissant à la production d'erreurs. Les occurrences de biais humains comprennent les stéréotypes, la perception sélective, l'effet d'entraînement, l'amorçage, la prédisposition à l'affirmation, le biais de sélection d'observation et la fausse notion du spéculateur. 

Dans de nombreux cas, il est essentiel d'éviter de tels biais pour les systèmes de ML afin de prendre des décisions rationnelles et optimales. Cela rendra les systèmes de ML plus pragmatiques que les humains si nous parvenons à déduire les biais humains et à les rectifier. Cela sera particulièrement utile dans les systèmes basés sur HITL. En utilisant les tests de biais, trois types de biais humains peuvent être identifiés et traités pour maintenir la prise de décision du système ML de telle sorte qu'elle soit exempte de biais humains. Ces trois biais humains sont les suivants:

* Biais d'interaction Lorsqu'un système de ML reçoit un ensemble de données contenant des entrées d'un type particulier, un biais d'interaction est introduit qui empêche l'algorithme de reconnaître d'autres types d'entrées. Ce type de biais peut être identifié dans les tests d'inférence pour les modèles entraînés. Des méthodes telles que SHAP et PFI peuvent être utiles pour identifier le biais de caractéristiques.

* Biais latent Un biais latent se produit lorsque plusieurs exemples de l'ensemble d'apprentissage ont une caractéristique qui se démarque. En fait, ceux qui n'ont pas cette caractéristique ne sont pas reconnus par l'algorithme. Par exemple, récemment, l'algorithme d'Amazon HR pour sélectionner des personnes en fonction des candidatures à des postes au sein de l'entreprise a montré un biais contre les femmes, la raison en étant un biais latent

* Biais de sélection Le biais de sélection est introduit dans un algorithme lorsque la sélection des données à analyser n'est pas correctement randomisée. Par exemple, dans la conception d'un système de reconnaissance faciale performant, il est essentiel d'inclure tous les types possibles de structures et de formes faciales et de tous les échantillons ethniques et géographiques, afin d'éviter les biais de sélection. Le biais de sélection peut être identifié par des méthodes telles que SHAP ou PFI pour observer le biais des caractéristiques du modèle.


## Politique optimale 

Dans le cas de l'apprentissage par renforcement humain, le but du système est de maximiser les récompenses de l'action dans l'état actuel. Afin de maximiser les récompenses des actions, une politique optimale peut être utilisée comme mesure pour évaluer le système. La politique optimale est la politique où l'action qui maximise la récompense/le retour de l'état actuel est choisie. La politique optimale est la métrique ou l'état idéal pour qu'un système fonctionne au mieux. Dans un système basé sur l'apprentissage par renforcement humain, un opérateur ou un enseignant humain définit la politique optimale car l'objectif du système est d'atteindre des performances de niveau humain.


##  Taux d'automatisation 

L'automatisation est le processus de production automatique de biens ou d'exécution d'une tâche à l'aide de robots ou d'algorithmes sans assistance humaine directe. Le niveau d'automatisation d'un système ML peut être calculé en utilisant le taux d'automatisation de l'ensemble des tâches. Il s'agit essentiellement du pourcentage de tâches entièrement automatisées par le système, et ces tâches ne nécessitent aucune assistance humaine. Il montre quel pourcentage de tâches sont entièrement automatisées sur toutes les tâches. Par exemple, AlphaGo, de DeepMind, a atteint une automatisation à 100 % pour fonctionner seul pour vaincre les joueurs champions du monde humains


## Taux de risque : 

la probabilité qu'un modèle de ML exécute des erreurs est appelée taux d'erreur. Le taux d'erreur est calculé en fonction des performances du modèle pour les systèmes de production. Plus le taux d'erreur est bas, meilleur c'est pour un système ML. L'objectif d'un humain dans la boucle est de réduire le taux d'erreur et d'apprendre au modèle ML à fonctionner de manière optimale

## Méthodes de test  de production

Comme il existe diverses entreprises en activité, il en va de même pour les différents types de systèmes de production au service de ces entreprises. Dans cette section, nous examinons les diérents types de systèmes ou de configurations de production couramment utilisés et comment les tester.


## Tests par lots 

Les tests par lots valident votre modèle en effectuant des tests dans un environnement différent de son environnement d'entraînement. Les tests par lots sont effectués sur un ensemble d'échantillons de données pour tester l'inférence du modèle à l'aide de métriques de choix, telles que la précision, le RMSE ou le score f1. Les tests par lots peuvent être effectués dans différents types de calculs, par exemple, dans le cloud, ou sur un serveur distant ou un serveur de test. Le modèle est généralement utilisé comme fichier sérialisé, et le fichier est chargé en tant qu'objet et déduit des données de test


## Tests A/B

Vous aurez sûrement déjà croisé les tests A/B. Il est souvent utilisé dans la conception de services (sites Web, applications mobiles, etc.) et pour évaluer les campagnes marketing. Par exemple, il est utilisé pour évaluer si un changement spécifique dans la conception ou l'adaptation du contenu à un public spécifique a un impact positif sur les mesures commerciales telles que l'engagement des utilisateurs, le taux de clics ou le taux de vente. Une technique similaire est appliquée pour tester les modèles de ML à l'aide de tests A/B. Lorsque les modèles sont testés à l'aide de tests A/B, le test répond à des questions importantes telles que les suivantes : 

* Le nouveau modèle B fonctionne-t-il mieux en production que le modèle A actuel ?

* Lequel des candidats des deux modèles fonctionne le mieux en production pour générer des indicateurs commerciaux positifs ?


Pour évaluer les résultats des tests A/B, des techniques statistiques sont utilisées en fonction de l'entreprise ou des opérations pour déterminer quel modèle sera le plus performant en production. Les tests A/B sont généralement effectués de cette manière, et les données en temps réel ou en direct sont fragmentées ou divisées en deux ensembles, l'ensemble A et l'ensemble B. Les données de l'ensemble A sont acheminées vers l'ancien modèle et les données de l'ensemble B sont acheminées vers le nouveau modèle. Afin d'évaluer si le nouveau modèle (modèle B) fonctionne mieux que l'ancien modèle (modèle A), diverses techniques statistiques peuvent être utilisées pour évaluer les performances du modèle (par exemple, exactitude, précision, rappel, f-score et RMSE) , selon le cas d'utilisation ou les opérations commerciales. En fonction d'une analyse statistique des performances du modèle en corrélation avec les mesures commerciales (un changement positif dans les mesures commerciales), la décision est prise de remplacer le nouveau modèle par l'ancien ou de déterminer quel modèle est le meilleur


Les tests A/B sont effectués méthodiquement à l'aide de tests d'hypothèses statistiques, et cette hypothèse valide les deux faces d'une pièce : l'hypothèse nulle et l'hypothèse alternative. L'hypothèse nulle affirme que le nouveau modèle n'augmente pas la valeur moyenne des métriques commerciales de surveillance. L'hypothèse alternative affirme que le nouveau modèle améliore la valeur moyenne des mesures commerciales de surveillance. En fin de compte, les tests A/B sont utilisés pour évaluer si le nouveau modèle génère une augmentation significative des métriques commerciales spécifiques. Il existe différents types de tests A/B, en fonction des cas d'utilisation et des opérations commerciales, par exemple, le test Z, le test G (je recommande de connaître ceux-ci et d'autres) et d'autres. Choisir le bon test A/B et les bonnes mesures à évaluer peut être une solution gagnant-gagnant pour votre entreprise et vos opérations de ML

## Stage test ou shadow test

Avant de déployer un modèle pour la production, ce qui conduirait ensuite à prendre des décisions commerciales, il peut être utile de répliquer un environnement de type production (environnement intermédiaire) pour tester les performances du modèle. Ceci est particulièrement important pour tester la robustesse du modèle et évaluer ses performances sur des données en temps réel. Cela pourrait être facilité en déployant la branche de développement ou un modèle à tester sur un serveur intermédiaire et en déduisant les mêmes données que le pipeline de production. Le seul inconvénient ici sera que les utilisateurs finaux ne verront pas les résultats de la branche de développement ou que les décisions commerciales ne seront pas prises dans le serveur intermédiaire. Les résultats de l'environnement de mise en scène seront évalués statistiquement à l'aide de métriques appropriées (par exemple, exactitude, précision, rappel, f-score et RMSE) pour déterminer la robustesse et les performances du modèle en corrélation avec les métriques commerciales



## Tests dans CI/CD

La mise en œuvre de tests dans le cadre de pipelines CI/CD peut être gratifiant en termes d'automatisation et d'évaluation (sur la base de critères définis) des performances du modèle. Les pipelines CI/CD peuvent être configurés de plusieurs manières en fonction des opérations et de l'architecture en place, par exemple :

* En cas d'exécution réussie d'un pipeline ML, les pipelines CI/CD peuvent déclencher le test A/B d'un nouveau modèle dans l'environnement de pré-production.

* Lorsqu'un nouveau modèle est formé, il est avantageux de configurer un ensemble de données distinct du test défini pour mesurer ses performances par rapport à des métriques appropriées, et cette étape peut être entièrement automatisée.

* Les pipelines CI/CD peuvent déclencher périodiquement des pipelines ML à une heure définie dans la journée pour former un nouveau modèle, qui utilise des données en direct ou en temps réel pour former un nouveau modèle ou affiner un modèle existant.

* Les pipelines CI/CD peuvent surveiller les performances du modèle ML du modèle déployé en production, et cela peut être déclenché ou géré à l'aide de déclencheurs temporels ou manuels (initiés par les membres de l'équipe responsables de l'assurance qualité).

* Les pipelines CI/CD peuvent provisionner deux environnements intermédiaires ou plus pour effectuer des tests A/B sur des ensembles de données uniques afin d'effectuer des tests plus diversifiés et plus complets


Il s'agit d'une variété de scénarios, et selon les exigences, les pipelines CI/CD offrent divers flux de travail et opérations adaptés aux besoins de l'entreprise et aux exigences technologiques. La sélection d'une architecture et d'un processus CI/CD efficaces peut augmenter les opérations technologiques et les performances globales de l'équipe. Les tests CI/CD peuvent augmenter et automatiser les tests sur de grandes longueurs

## Pourquoi packager des modèles de ML ?

MLOps permet une approche systématique pour former et évaluer des modèles. Une fois les modèles formés et évalués, les prochaines étapes consistent à les mettre en production. Comme nous le savons, le ML ne fonctionne pas comme l'ingénierie logicielle traditionnelle, qui est de nature déterministe et où un morceau de code ou un module est importé dans le système existant et cela fonctionne. Les solutions d'ingénierie ML sont non déterministes et impliquent de servir des modèles ML pour faire des prédictions ou analyser des données. Afin de servir les modèles, ils doivent être emballés dans des artefacts logiciels pour être expédiés vers les environnements de test ou de production. Habituellement, ces artefacts logiciels sont regroupés dans un fichier ou un groupe de fichiers ou de conteneurs. Cela permet au logiciel d'être indépendant de l'environnement et du déploiement. Les modèles ML doivent être emballés pour les raisons suivantes :


### Portabilité 

Packaging  des modèles ML dans des artefacts logiciels leur permet d'être expédiés ou transportés d'un environnement à un autre. Cela peut être fait en expédiant un fichier ou un tas de fichiers ou un conteneur. Dans tous les cas, nous pouvons transporter les artefacts et reproduire le modèle dans diverses configurations. Par exemple, un modèle packagé peut être déployé dans une machine virtuelle ou une configuration sans serveur


## L'inférence 

Inference ML est un processus qui implique le traitement de données en temps réel à l'aide de modèles ML pour calculer une sortie, par exemple, une prédiction ou un score numérique. Le but de l'empaquetage des modèles ML est de pouvoir servir les modèles ML en temps réel pour l'inférence ML. Un packaging de modèle ML efficace (par exemple, un modèle ou un conteneur sérialisé) peut faciliter le déploiement et servir le modèle pour faire des prédictions et analyser les données en temps réel ou par lots

## Interopérabilité 

L'interopérabilité des modèles ML est la capacité de deux ou plusieurs modèles ou composants à échanger des informations et à utiliser les informations échangées afin d'apprendre ou de s'adapter les uns aux autres et d'effectuer des opérations avec efficacité. Les informations échangées peuvent prendre la forme de données ou d'artefacts logiciels ou de paramètres de modèle. Ces informations permettent aux modèles d'affiner, de se recycler ou de s'adapter à divers environnements à partir de l'expérience d'autres artefacts logiciels afin d'être performants et efficaces. L'empaquetage des modèles ML est la base pour permettre l'interopérabilité des modèles ML



## Agnostique de déploiement 

L'empaquetage des modèles ML dans des artefacts logiciels tels que des fichiers ou des conteneurs sérialisés permet aux modèles d'être expédiés et déployés dans divers environnements d'exécution, tels qu'une machine virtuelle, un environnement sans serveur de conteneur, un service de streaming, des microservices ou un traitement par lots prestations de service. Il ouvre des opportunités de portabilité et d'agnosticité de déploiement en utilisant les mêmes artefacts logiciels dans lesquels un modèle ML est packagé

Les modèles ML peuvent être pakager de différentes manières en fonction des exigences commerciales et techniques et selon les opérations de ML. Les modèles ML peuvent être pakager et expédiés de trois manières, comme indiqué dans les sous-sections suivantes


## Fichiers sérialisés 

La sérialisation est un processus vital pour l'empaquetage d'un modèle ML car il permet la portabilité, l'interopérabilité et l'inférence de modèle. La sérialisation est la méthode de conversion d'un objet ou d'une structure de données (par exemple, des variables, des tableaux et des tuples) en un artefact stockable, par exemple, en un fichier ou un tampon mémoire qui peut être transporté ou transmis (à travers un ordinateur réseaux). L'objectif principal de la sérialisation est de reconstruire le fichier sérialisé dans sa structure de données précédente (par exemple, un fichier sérialisé dans une variable de modèle ML) dans un environnement différent. De cette façon, un modèle ML nouvellement formé peut être sérialisé dans un fichier et exporté dans un nouvel environnement où il peut être désérialisé de nouveau dans une variable de modèle ML ou une structure de données pour l'inférence ML. Un fichier sérialisé n'enregistre ni n'inclut aucune des méthodes ou implémentations précédemment associées. Il enregistre uniquement la structure des données telle qu'elle est dans un artefact stockable tel qu'un fichier.

Ces défis peuvent être évités en standardisant certains processus ou modules et les conteneurs sont un excellent moyen de standardiser les modèles de ML et les modules logiciels. 

Un conteneur est une unité standard de logiciel composée de code et de toutes ses dépendances. Il permet le fonctionnement rapide et fiable des applications d'un environnement informatique à un autre. Il permet au logiciel d'être indépendant de l'environnement et du déploiement. 

Les conteneurs sont gérés et orchestrés par Docker. Docker est devenu un standard de l'industrie pour le développement et l'orchestration des conteneurs. Docker est un outil open source (https://opensource.com/resources/what-open-source). Il a été développé pour faciliter la création, le déploiement et l'exécution d'applications à l'aide de conteneurs. En utilisant des conteneurs, un développeur peut empaqueter une application avec ses composants et modules, tels que des fichiers, des bibliothèques et d'autres dépendances, et la déployer comme un seul package. 

Les conteneurs sont un moyen fiable d'exécuter des applications à l'aide d'un système d'exploitation Linux avec des paramètres personnalisés. Les conteneurs Docker sont construits à l'aide de Dockerfiles, qui sont utilisés pour conteneuriser une application. 

Après la création d'une image Docker, un conteneur Docker est créé. Un conteneur Docker est une application exécutée avec des paramètres personnalisés orchestrés par le développeur.

Un Dockerfile, une image Docker et un conteneur Docker sont des composants fondamentaux pour la création et l'exécution de conteneurs. Ceux-ci sont chacun décrits ici :

* Dockerfile : un Dockerfile est un document texte contenant un ensemble de commandes Docker commandées par le développeur pour créer une image Docker. Docker est capable de lire le fichier Docker et de créer une image Docker.

* Image Docker : il s'agit d'une collection séquentielle de paramètres d'exécution à utiliser dans une collection de systèmes de fichiers racine dans un conteneur pendant l'exécution. Les images Docker sont comme un instantané de conteneurs. Les conteneurs sont construits à partir d'images Docker

* Conteneur Docker : les conteneurs sont construits à partir d'images Docker. Un conteneur est une instance d'exécution d'une image Docker

## Génération et déploiement de microservicesLes 

microservices permettent la collecte de services déployables indépendamment. Chacun de ces services est hautement maintenable, testable et faiblement couplé. Les microservices sont orchestrés par une architecture organisée autour de capacités métier pour permettre à un système de répondre aux besoins de l'entreprise. Par exemple, Spotify est passé d'un système complexe monolithique à un système basé sur des microservices. Cela a été fait en divisant le système complexe en services fragmentés, avec des objectifs spécifiques tels qu'un moteur de recherche, le balisage du contenu, la classification du contenu, l'analyse du comportement des utilisateurs pour un moteur de recommandation et des listes de lecture générées automatiquement. 

Les microservices fragmentés sont désormais développés par une équipe dédiée. Chaque microservice est isolé et moins dépendant les uns des autres. De cette façon, il est plus facile à développer et à maintenir. L'entreprise peut être cohérente avec le service client et s'améliorer continuellement sans interrompre le service.


En règle générale, un microservice est généré en adaptant des fichiers sérialisés dans une image Docker conteneurisée. Ces images Docker peuvent ensuite être déployées et orchestrées dans n'importe quel environnement pris en charge par Docker. Le déploiement et la gestion des images Docker peuvent être effectués à l'aide d'outils de gestion de conteneurs, tels que Kubernetes. Docker permet une portabilité et une interopérabilité extrêmes, les images Docker peuvent être facilement déployées sur n'importe quel service cloud populaire, tel que Google Cloud, Azure ou AWS. Les images Docker peuvent être déployées et gérées sur n'importe quel serveur d'entreprise ou centre de données Docker ou environnement en temps réel tant qu'il prend en charge Docker


Les microservices peuvent être servis dans un format d'API REST, et c'est un moyen courant de servir des modèles de ML. Certains frameworks Python, tels que Flask, Django et FastAPI, sont devenus populaires pour permettre aux modèles ML de servir de microservices d'API REST. Pour faciliter des opérations système robustes et évolutives, les développeurs de logiciels peuvent se synchroniser avec des microservices Dockerisés via une API REST. Pour orchestrer les déploiements de microservices basés sur Docker sur une infrastructure prise en charge par Kubernetes, Kubeflow est une bonne option. Il est indépendant du cloud et peut être exécuté sur site ou sur des machines locales. En plus de cela, Kubeflow est basé sur Kubernetes, mais conserve les détails et les difficultés de Kubernetes dans une boîte. Kubeflow est un moyen robuste de servir un modèle. ## Connexion à l'espace de travail et importation des artefacts de modèle

Tout d'abord, nous importons les packages requis, nous nous connectons à l'espace de travail ML à l'aide de la fonction Workspace(), puis nous téléchargeons le scaler et le modèle sérialisés pour effectuer des prédictions. Scaler sera utilisé pour mettre à l'échelle les données d'entrée dans la même échelle de données que celle utilisée pour l'apprentissage du modèle. Le fichier modèle est sérialisé au format ONNX. Les fichiers Scaler et Model sont importés à l'aide de la fonction Model()

```python
import pandas as pd
import numpy as np
import warnings
import pickle
from math import sqrt
warnings.filterwarnings('ignore')
from azureml.core.run import Run
from azureml.core.experiment import Experiment
from azureml.core.workspace import Workspace
from azureml.core.model import Model
from azureml.core.authentication import ServicePrincipalAuthentication


# Connect to Workspace
ws = Workspace.from_config()
print(ws)


scaler = Model(ws,'scaler').download(exist_ok=True)

svc_model = Model(ws,'support-vector-classifier').download(exist_ok=True)
```

## Chargement des artefacts de modèle pour l'inférence

Nous ouvrons et chargeons les fichiers Scaler et Model dans des variables pouvant être utilisées pour l'inférence de modèle ML. Le scaler est lu et chargé dans une variable à l'aide de pickle, et l'environnement d'exécution ONNX est utilisé pour charger le fichier ONNX à l'aide d'InferenceSession() pour faire des prédictions de modèle ML comme suit :

```python
with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)

  
# Compute the prediction with ONNX Runtime
import onnxruntime as rt
import numpy
sess = rt.InferenceSession("svc.onnx")


input_name = sess.get_inputs()[0].name
label_name = sess.get_outputs()[0].name

```

## Inférence de modèle ML

Pour effectuer une inférence de modèle ML, mettez les données de test à l'échelle et configurez-les pour l'inférence à l'aide de la méthode fit_transform(). À présent, effectuez une inférence sur les données de test en utilisant la session ONNX et exécutez sess.run() en transmettant les données d'entrée, test_data, au format float 32. Enfin, imprimez les résultats de l'inférence de modèle :

```python
test_data = np.array([34.927778, 0.24, 7.3899, 83, 16.1000, 1])

# Scale data
test_data = scaler.fit_transform(test_data.reshape(1, 6))


pred_onx = sess.run([label_name], {input_name: test_data.astype(numpy.float32)})[0]

pred_onx[0]
```

Avec ces étapes, nous avons téléchargé avec succès le modèle sérialisé, l'avons chargé dans une variable et effectué une inférence sur un échantillon de données de test. Le résultat attendu du code de bloc est la valeur 1
