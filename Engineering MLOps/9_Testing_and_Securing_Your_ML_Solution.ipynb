En plus d'effectuer des tests réguliers de développement de logiciels, tels que des tests unitaires, des tests d'intégration, des tests système et des tests d'acceptation, les solutions ML nécessitent des tests supplémentaires car les données et les modèles ML sont impliqués. ## Test des données : 

l'objectif du test des données est de s'assurer que les données sont d'une qualité suffisamment élevée pour l'apprentissage du modèle ML. Plus la qualité des données est bonne, meilleurs sont les modèles entraînés pour les tâches données. Alors comment évaluer la qualité des données ? Cela peut être fait en inspectant les cinq facteurs suivants des données : 

* Accuracy 

* Complétude (aucune valeur manquante)
* Cohérence (en termes de format et de volume de données attendus)

* Pertinence (les données doivent répondre aux besoins et aux exigences prévus) * Actualité (les données les plus récentes ou à jour)


Sur la base de ces facteurs, si une entreprise peut gérer la qualité des données de chaque ensemble de données lors de leur réception ou de leur création, la qualité des données est garantie. Voici quelques étapes que votre équipe ou entreprise peut utiliser comme mesures d'assurance qualité pour vos données : 


1. Catalogage méticuleux des données et contrôle des données entrantes : une combinaison de catalogage des données (pour documenter et stocker les données dans le format ou le modèle requis) et de fonctions de contrôle peut garantir une haute qualité des données entrantes. Le catalogage et le contrôle des données peuvent être effectués en surveillant les facteurs de données tels que les formats et les modèles de données, les distributions de valeurs et les anomalies, l'exhaustivité et la cohérence peuvent aider à fournir une bonne qualité des données entrantes.

2. Conserver soigneusement les pipelines de données pour éviter les données en double : lorsque les données dupliquées sont fabriquées à partir de la même source de données et en utilisant la même logique par différentes personnes, il peut devenir compliqué de gérer les lignées, l'authenticité et l'intégrité des données. Cela peut produire des effets en cascade sur plusieurs systèmes ou bases de données. Il est préférable d'éviter autant que possible de dupliquer les données.

3. Gouvernance des données avec intégrité forcée : Dans le monde d'aujourd'hui, le maintien de l'intégrité des données est devenu crucial. Ne pas avoir la mentalité de faire respecter l'intégrité des données peut être coûteux pour une organisation. Les données pourraient éventuellement devenir incomplètes, retardées ou obsolètes, ce qui entraînerait de graves problèmes de qualité des données.

4. Maintien de la traçabilité et du lignage de bout en bout : le lignage et la traçabilité des données peuvent être obtenus grâce à l'utilisation intelligente des métadonnées et des données elles-mêmes. En utilisant les deux, nous pouvons documenter des informations critiques telles que des clés uniques pour chaque ensemble de données, ajouter un horodatage à chaque enregistrement et consigner les modifications des données. S'assurer que le lignage des données et la traçabilité de bout en bout sont activés peut nous donner la possibilité de reproduire des modèles et de déboguer les erreurs et les pipelines

## Tests de modèles 

Les tests de modèles doivent couvrir les problèmes de serveur tels que les suivants :

* Évaluation de la précision ou de la métrique clé du modèle ML 

* Tests sur des points de données aléatoires
* Tests de perte ou de performances acceptables pour votre tâche 
* Tests unitaires de robustesse du modèle à l'aide de données réelles 

Ces tests peuvent être orchestrés en deux phases : pré-formation et post-formation. Faciliter ces tests dans le flux de travail peut produire des modèles robustes pour la production. Regardons quels tests pré-train et post-train peuvent être effectués par conception.

## Tests de pré-formation 

Des tests peuvent être effectués pour détecter les défauts avant de passer à l'étape de la formation. Ces défauts peuvent se trouver dans les données, les pipelines ou les paramètres. La figure 9.1 suggère d'exécuter des tests de pré-formation et de post-formation dans le cadre d'un flux de travail proposé pour développer des modèles de haute qualité :

image.png

Voici quelques façons de détecter et d'éviter les failles de pré-entraînement à l'aide de tests de pré-entraînement :

* Éliminer la dette du pipeline de données en gérant les fuites de données, les cas extrêmes et en optimisant pour rendre le pipeline efficace en termes de temps et de ressources

* S'assurer la forme de la sortie de votre modèle correspond aux étiquettes de votre ensemble de données

* Examiner les plages de sortie pour s'assurer qu'elles correspondent à nos attentes (par exemple, vérifier que la sortie d'un modèle de classification est une distribution avec des probabilités de classe qui totalisent 1)

* Examiner vos ensembles de données de formation et de validation pour les fuites d'étiquettes 

* S'assurer que le pipeline ETL génère ou récupère les données dans le format requis

Les tests de pré-entraînement n'ont pas besoin de paramètres pour s'exécuter, mais ils peuvent être très utiles pour détecter les bogues avant d'exécuter l'entraînement du modèle

## Tests de post-formation 

Les tests de post-formation nous permettent d'étudier les performances du modèle et la logique derrière les prédictions du modèle et de prévoir les éventuelles failles du modèle avant de déployer le modèle en production. Les tests post-formation nous permettent de détecter les défauts de performance et de fonctionnalité du modèle. Les tests post-formation impliquent un test d'évaluation des performances du modèle, un test d'invariance et un test de fonctionnalité minimale. Voici une lecture recommandée pour plus d'informations sur les tests post-formation : Beyond Accuracy : Behavioral Testing of NLP Models with CheckList (https://homes.cs.washington.edu/~marcotcr/acl20_checklist.pdf)

## Types d'attaques

Nous explorerons certaines des attaques les plus courantes contre les systèmes de ML. À un niveau élevé, les attaques de pirates peuvent être divisées en quatre catégories : empoisonnement, attaques par saisie et évasion, ingénierie inverse et attaques par porte dérobée. Voyons comment les attaquants parviennent à infiltrer les systèmes de ML via ces attaques.


## Empoisonnement

Un pirate informatique ou un attaquant cherche à compromettre un modèle d'IA lors d'une attaque par empoisonnement. Les attaques par empoisonnement peuvent survenir à n'importe quelle étape (entraînement, déploiement ou inférence en temps réel). Ils se produisent généralement dans l'apprentissage et l'inférence. Voyons comment les attaques par empoisonnement sont mises en œuvre de trois manières typiques :

* Empoisonnement de l'ensemble de données : les ensembles de données d'entraînement contiennent les connaissances sur lesquelles le modèle est entraîné. Un attaquant peut manipuler ces connaissances en infiltrant le jeu de données d'apprentissage. Ici, l'attaquant introduit des données mal étiquetées ou incorrectes dans l'ensemble de données d'apprentissage, et avec cela, l'ensemble du processus d'apprentissage est déformé. C'est un moyen direct d'empoisonner un modèle. Les ensembles de données d'entraînement peuvent être empoisonnés pendant les phases de collecte et de conservation des données, et il peut être difficile de le remarquer ou de le détecter car les ensembles de données d'entraînement peuvent provenir de plusieurs sources, peuvent être volumineux et aussi car l'attaquant peut s'infiltrer dans les distributions de données.

* L'empoisonnement de l'algorithme se produit lorsqu'un attaquant se mêle de l'algorithme utilisé pour entraîner le modèle. Cela peut être aussi simple que d'intégrer des hyperparamètres ou de jouer avec l'architecture de l'algorithme. Par exemple, prenons l'apprentissage fédéré (qui vise à préserver la confidentialité des données des individus) où la formation du modèle est effectuée sur plusieurs sous-ensembles de données privées (telles que les données de santé de plusieurs hôpitaux tout en préservant les informations confidentielles des patients). Plusieurs modèles sont dérivés de chaque sous-ensemble, puis combinés pour former un modèle final. Pendant ce temps, un attaquant peut manipuler n'importe quel sous-ensemble des données et influencer le modèle final résultant. L'attaquant peut également créer un faux modèle à partir de fausses données et le concaténer avec des modèles produits à partir d'un entraînement sur plusieurs sous-ensembles de données privées pour produire un modèle final qui s'écarte de l'exécution efficace de la tâche, ou sert les motivations de l'attaquant.

* Modèle l'empoisonnement se produit lorsqu'un attaquant remplace un modèle déployé par un modèle alternatif. Ce type d'attaque est identique à une cyberattaque typique où les fichiers électroniques contenant le modèle pourraient être modiés ou remplacés.


## Attaque d'entrée et évasion 

Une attaque d'entrée ou d'évasion se produit lorsque l'attaquant modifie l'entrée du système ML de manière à provoquer un dysfonctionnement du système (ou à donner une mauvaise prédiction). Ces perturbations ou changements peuvent être difficiles à détecter car ces changements sont très subtils ou petits. Par exemple, les attaques d'entrée sont courantes pour les algorithmes de vision par ordinateur. 

Cela peut être fait en changeant simplement quelques pixels dans l'image d'entrée. En conséquence, le système peut identifier une image d'une manière qu'il n'est pas censé faire ou faire une mauvaise prédiction. De tels petits changements peuvent efficacement manipuler la prédiction, entraînant des actions erronées prises par le système. En conséquence, le système ML se comporte comme il se doit, tandis que la sortie est manipulée. 

Les systèmes ML sont très sujets aux attaques d'entrée. Par conséquent, disposer d'un détecteur d'anomalies pour surveiller les données entrantes peut être très pratique pour éviter de telles perturbations dans les données entrantes. Indépendamment des données d'entrée, la majorité des modèles de classification choisissent une classe valide à partir de leur apprentissage. Une autre façon de protéger le système ML consiste à prétraiter les entrées avec un modèle binaire proxy qui vous indique, par exemple, si une image d'entrée est celle d'une personne ou d'un animal avant d'envoyer cette image au classeur d'images final.


## Reverse engineering

Pour un utilisateur d'un système d'IA, il peut s'agir d'une boîte noire ou opaque. Il est courant dans les systèmes d'IA d'accepter des entrées pour générer des sorties sans révéler ce qui se passe à l'intérieur (en termes à la fois de logique et d'algorithme). Les ensembles de données d'entraînement, qui contiennent effectivement toutes les connaissances du système entraîné, sont également généralement gardés confidentiels. Cela, en théorie, empêche un étranger de prédire pourquoi des sorties particulières sont produites ou ce qui se passe à l'intérieur du système d'IA en termes d'algorithme, de données d'entraînement ou de logique. 

Cependant, dans certains cas, ces systèmes peuvent être sujets à l'ingénierie inverse. L'objectif de l'attaquant ou des pirates dans les attaques d'ingénierie inverse est de reproduire le modèle d'origine déployé en tant que service et de l'utiliser à leur avantage. Dans un article intitulé Attaques d'extraction de modèle contre les réseaux neuronaux récurrents (https://arxiv.org/pdf/ 2002.00123.pdf), publié en février 2020, les chercheurs ont mené des expériences sur des attaques d'extraction de modèle contre un RNN et un LSTM entraînés avec des ensembles de données académiques accessibles au public.

 Les chercheurs reproduisent efficacement la fonctionnalité d'un système ML via une attaque d'extraction de modèle. Ils démontrent qu'une attaque d'extraction de modèle avec une grande précision peut être extraite efficacement, principalement en répliquant ou en configurant une fonction de perte ou une architecture à partir du modèle cible. ## Attaques de porte dérobée 

Dans les attaques de porte dérobée, les attaques peuvent intégrer des modèles de leur choix dans le modèle dans les étapes d'entraînement ou d'inférence et déduire le modèle déployé à l'aide d'entrées pré-organisées pour produire des sorties ou des déclencheurs inattendus pour le système ML. 

Par conséquent, les attaques de porte dérobée peuvent se produire à la fois dans les phases d'entraînement et d'inférence, tandis que les attaques d'évasion et d'empoisonnement peuvent se produire en une seule phase pendant l'entraînement ou l'inférence. le modèle étudiant peut apprendre à pirater certaines portes dérobées du modèle enseignant à l'aide de l'apprentissage par transfert. le modèle ou le système. En outre, les attaques de porte dérobée peuvent viser à dégrader les performances, à épuiser ou à rediriger des ressources pouvant entraîner une défaillance du système, ou à tenter d'introduire un comportement et des sorties particuliers du système d'IA.
