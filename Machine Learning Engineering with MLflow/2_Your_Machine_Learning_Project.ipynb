## Énoncé du problème 

Comprendre le problème que nous résolvons est très important avant d'essayer quoi que ce soit d'autre. Dénir un énoncé de problème pour le problème en question est un moyen clair de dénir votre problème. Voici des exemples raisonnables d'énoncé de problème d'apprentissage automatique :

* Prédisez si le marché boursier d'un ticker spécifique augmentera demain.
* Prédisez si nous avons un chat dans une image

Dans cette partie du processus, nous spécifions le type spécifique de problème d'apprentissage automatique que nous visons à résoudre. Les types de problèmes suivants sont courants :

* Classication : ce type de problème nécessite que vous prédisiez une étiquette ou une classe en tant que sortie de votre modèle, par exemple, en classant si le texte d'un e-mail est un spam ou non. Il peut s'agir d'une classification binaire ou d'une classification multiclasse. Un bon exemple de variante multiclasse consiste à classer les chiffres en fonction de l'écriture manuscrite. 

* Régression : c'est si vous avez besoin, par exemple, d'avoir un modèle prédisant une valeur numérique à partir d'un ensemble de données d'apprentissage. De bons exemples sont la prévision de la température basée sur les caractéristiques atmosphériques et la prévision de la valeur exacte en dollars d'un stock donné. 

* Regroupement : consiste à découvrir des regroupements naturels de données lorsque vous n'avez pas d'étiquettes pour entraîner des modèles. Il regroupe des points de données similaires dans les mêmes groupes à l'aide d'une métrique de distance. Le clustering peut être utilisé pour identifier les transactions frauduleuses car elles n'appartiendront pas à un regroupement existant.

* Modèle génératif : il s'agit d'un nouveau type d'apprentissage automatique dans lequel de nouvelles données sont générées sur la base des données existantes et sont statistiquement similaires aux données d'entrée . Il est largement utilisé dans les modèles de langage moderne tels que GPT-3 d'OpenAI.

* Apprentissage par renforcement : un type d'apprentissage automatique dans lequel l'agent/le modèle interagit avec l'environnement pour apprendre un comportement optimal afin de maximiser la récompense. Une application célèbre est l'agent AlphaGo de Google DeepMind, qui a pu surpasser le meilleur joueur humain au jeu de société Go. Une application très importante consiste à former un agent boursier automatisé en utilisant la rentabilité comme récompense

## Dénition du succès et de l'échec

Obtenir la définition du succès d'une entreprise est très important pour mettre votre problème en perspective. Par exemple, pour le cas du marché boursier, ce qui suit peut être décrit comme un résultat souhaitable : « En utilisant le modèle de ce projet, nous aimerions améliorer la rentabilité jusqu'à 56% de nos opérations commerciales quotidiennes car elles sont actuellement pro table dans 50% des transactions quotidiennes."

Donc, si nous avons des transactions réussies seulement 30% du temps, le modèle est clairement un échec. Cela dépendra généralement de critères commerciaux. Avoir une métrique commerciale comme définition du succès, au lieu d'une métrique technique telle que l'exactitude, la précision ou le rappel, aide à maintenir la solution alignée sur les objectifs tangibles d'une organisation.

Par exemple, la sortie d'un modèle de régression sera un nombre spécifique (par exemple, 10,5 comme prévision de stock) et une classification renverra une étiquette (par exemple, vrai lorsqu'un chat est détecté) et un seuil probabiliste.Votre la sortie du modèle doit certainement être liée à la définition de votre résultat

1. Implémentez le modèle heuristique dans MLflow. Dans le bloc de code suivant, nous créons la classe RandomPredictor, un prédicteur sur mesure qui descend de la classe mlflow.pyfunc.Python Model. La méthode de prédiction principale renvoie un nombre aléatoire compris entre 0 et 1 :

```python
import mlflow

class RandomPredictor(mlflow.pyfunc.PythonModel):  
  def __init__(self):    
    pass  
  
  def predict(self, context, model_input):

    return model_input.apply(lambda column: random.randint(0,1))
```

Nous utilisons une fonctionnalité spécifique de création d'un modèle personnalisé dans MLflow ; plus de détails sur les modèles personnalisés peuvent être trouvés sur https://mlflow.org/docs/latest/python_api/mlflow.pyfunc.html#pyfunc-create-custom


2. Enregistrez le modèle dans MLflow. Le bloc de code suivant enregistre le modèle avec le nom random_model d'une manière qui peut être récupérée ultérieurement. Il s'enregistre dans le registre MLflow du système de fichiers local :

```python
model_path = "random_model"

baseline_model = RandomPredictor()

mlflow.pyfunc.save_model(path=model_path, python_model=random_model)
```
