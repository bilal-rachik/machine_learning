## Comprendre la valeur d'un atelier de science des données 

Un atelier de science des données est un environnement permettant de standardiser les outils et les pratiques d'apprentissage automatique d'une organisation, permettant une intégration et un développement rapides de modèles et d'analyses. 

L'une des fonctions essentielles d'ingénierie de l'apprentissage automatique consiste à soutenir les praticiens de la science des données avec des outils qui renforcent et accélèrent leurs activités quotidiennes. 

Dans une équipe de science des données, la capacité de tester rapidement plusieurs approches et techniques est primordiale. Chaque jour, de nouvelles bibliothèques et outils open source sont créés. Il est courant qu'un projet ait besoin de plus d'une dizaine de bibliothèques pour tester un nouveau type de modèle. 

Ces multitudes de bibliothèques, si elles ne sont pas assemblées correctement, peuvent provoquer des bogues ou des incompatibilités dans le modèle. Les données sont au centre d'un flux de travail de science des données. Il est essentiel de disposer d'ensembles de données propres pour développer et évaluer des modèles. Avec une abondance d'énormes ensembles de données, des outils spécialisés de Big Data sont nécessaires pour traiter les données. 

Les données peuvent apparaître dans plusieurs formats et vitesses pour l'analyse ou l'expérimentation, et peuvent être disponibles dans plusieurs formats et supports. Il peut être disponible via des fichiers, le cloud ou des interfaces de programmation d'applications (API) REpresentational State Transfer (REST)


La science des données est principalement un cra collaboratif; cela fait partie d'un flux de travail pour partager des modèles et des processus entre les membres de l'équipe. Invariablement, un point douloureux qui émerge de cette activité est la reproductibilité croisée des travaux de développement de modèles parmi les praticiens. 

Le scientifique des données A partage un script de formation d'un modèle qui suppose la version 2.6 d'une bibliothèque, mais le scientifique des données B utilise l'environnement de la version 2.8. Le traçage et la résolution du problème peuvent prendre des heures dans certains cas. 

Si ce problème se produit dans un environnement de production, il peut devenir extrêmement coûteux pour l'entreprise. Lors de l'itération, par exemple, sur un modèle, chaque exécution contient plusieurs paramètres qui peuvent être modifiés pour l'améliorer. Maintenir la traçabilité de quel paramètre a donné une mesure de performance spécifique, telle que la précision, par exemple, peut être problématique si nous ne stockons pas les détails de l'expérience de manière structurée. Revenir à un lot spécifique de paramètres qui ont produit un meilleur modèle peut être impossible si nous ne conservons que les derniers paramètres pendant la phase de développement du modèle. Le besoin d'itérer rapidement peut entraîner de nombreuses frustrations lors de la traduction du code du prototype vers un environnement de production, où il peut être exécuté de manière fiable. 

Par exemple, si vous développez un nouveau modèle commercial sur une machine Windows avec un accès facile aux unités de traitement graphique (GPU) pour l'inférence, le membre de votre équipe d'ingénierie peut décider de réutiliser l'infrastructure Linux existante sans accès GPU. Cela conduit à une situation où votre algorithme de production finit par prendre 5 heures et s'exécute localement en 30 secondes, ce qui a un impact sur le résultat final du projet. Il est clair qu'un département de science des données risque une douleur technique systémique si des problèmes liés à l'environnement et les outils ne sont pas abordés en amont. 


Pour résumer, nous pouvons énumérer les points principaux suivants, tels que décrits dans cette section :

* Frottements de reproductibilité
* La complexité de la gestion d'ensembles de données volumineux et variés
* Mauvaise gestion des paramètres d'expérimentation

* Drive entre les environnements locaux et de production

Un atelier de science des données résout les problèmes décrit dans cette section en créant un environnement structuré dans lequel un praticien de l'apprentissage automatique peut être habilité à développer et déployer ses modèles de manière fiable, avec une friction réduite. Un environnement sans friction permettra aux heures de développement de modèles très coûteuses de se concentrer sur le développement et l'itération de modèles, plutôt que sur la résolution de problèmes techniques d'outillage et de données. Après avoir exploré la motivation pour créer un atelier de science des données pour une équipe d'apprentissage automatique , nous allons ensuite commencer à concevoir l'atelier de science des données en fonction des points douloureux connus

## Créer votre propre atelier de science des données 

Afin de résoudre les problèmes courants liés au développement de modèles en science des données, comme décrit dans la section précédente, nous devons fournir aux scientifiques et aux praticiens des données un environnement standardisé dans lequel ils peuvent développer et gérer leur travail. Un atelier de science des données devrait vous permettre de démarrer rapidement un projet, et la disponibilité d'un environnement avec un ensemble d'outils et de cadres de démarrage permet aux scientifiques des données de démarrer rapidement un projet. centre de l'établi : ils doivent disposer d'une plate-forme fiable qui leur permet de développer et d'ajouter de la valeur à l'organisation, avec leurs modèles à portée de main. Afin de réfléchir à la conception de notre atelier de science des données et sur la base du diagramme de la figure 3.1, nous avons besoin des fonctionnalités de base suivantes dans notre atelier de science des données :

* Gestion des dépendances : avoir une gestion des dépendances intégrée à votre environnement local aide à gérer les problèmes de reproductibilité et la prévention des conflits de bibliothèque entre les différents environnements. Ceci est généralement réalisé en utilisant des gestionnaires d'environnement tels que Docker ou en ayant des cadres de gestion d'environnement disponibles dans votre langage de programmation. MLflow fournit cela grâce à la prise en charge des environnements basés sur Docker ou Conda.


* Gestion des données : la gestion des données dans un environnement local peut être complexe et intimidante si vous devez gérer d'énormes ensembles de données. Avoir une définition standardisée de la façon dont vous gérez les données dans vos projets locaux permet à d'autres de collaborer librement sur vos projets et de comprendre les structures disponibles.

* Gestion des modèles : avoir les différents modèles organisés et correctement stockés fournit une structure facile pour pouvoir travailler sur de nombreuses idées en même temps et persister celles qui ont du potentiel. MLflow aide à prendre en charge cela grâce à l'abstraction du format du modèle et au composant Model Registry pour gérer les modèles.

* Déploiement : avoir un environnement de développement aligné sur l'environnement de production où le modèle sera entretenu nécessite une délibération dans l'environnement local. L'environnement de production doit être prêt à recevoir un modèle d'un développeur de modèles, avec le moins de frictions possible. Ce flux de travail de déploiement fluide n'est possible que si l'environnement local est conçu correctement.

* Gestion de l'expérimentation : l'ajustement des paramètres est la chose la plus courante qu'un praticien de l'apprentissage automatique fait. Être capable de se tenir au courant des difiérentes versions et des paramètres spéciques peut rapidement devenir fastidieux pour le développeur de modèles.



Les équipes d'ingénierie en apprentissage automatique sont libres en termes de cas d'utilisation et de technologies que l'équipe qu'elles servent utilisera. Les étapes suivantes illustrent un bon flux de travail pour le développement avec un atelier de science des données :
